```python
def withdraw_cash(amount, available_notes, withdrawal_today):
    MAX_WITHDRAWAL = 50000
    DAILY_LIMIT = 100000
    DENOMINATIONS = [2000, 500, 100]
    
    if amount % 100 != 0:
        return "Invalid amount"
    if amount > MAX_WITHDRAWAL:
        return "Transaction Limit exceeded"
    if withdrawal_today + amount > DAILY_LIMIT:
        return "Daily Limit exceeded"
    
    notes_dispensed = {2000: 0, 500: 0, 100: 0}
    remaining_amount = amount
    
    for note in DENOMINATIONS:
        if remaining_amount >= note:
            num_notes = min(remaining_amount // note, available_notes.get(note, 0))
            notes_dispensed[note] = num_notes
            remaining_amount -= num_notes * note

    if remaining_amount > 0:
        return "Insufficient funds"
    

    for note in notes_dispensed:
        available_notes[note] -= notes_dispensed[note]

    return notes_dispensed


available_notes = {2000: 10, 500: 10, 100: 10}  
withdrawal_today = 0  

# Input and function call
amount = int(input('Enter the withdrawal amount: '))
result = withdraw_cash(amount, available_notes, withdrawal_today)

# Display the output
if isinstance(result, dict):
    print("Cash dispensed:")
    for note, count in result.items():
        if count > 0:
            print(f"{note} * {count}")
else:
    print(result)  
```

```python
current = input().strip()
target = input().strip()

rotations = 0
for i in range(len(current)):
    diff = abs(int(current[i]) - int(target[i]))
    rotations += min(diff, 10 - diff)

print(rotations)
```

```python
def minion_game(s):
    n = len(s)  
    kevin_score = 0
    stuart_score = 0
    vowels = 'AEIOU'
    
    for i in range(n):
        if s[i] in vowels:
            kevin_score += n - i  # Vowel player (Kevin)
        else:
            stuart_score += n - i  # Consonant player (Stuart)
    
    if kevin_score > stuart_score:
        print(f"Kevin {kevin_score}")
    elif stuart_score > kevin_score:
        print(f"Stuart {stuart_score}")
    else:
        print("Draw")

if __name__ == '__main__':
    s = input().strip().upper()
    minion_game(s)
```
