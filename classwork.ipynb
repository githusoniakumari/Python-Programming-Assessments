```python
class Parent:
    a=10
    b=20
    
    
    def __init__(self,c,d):
        self.c=c
        self.d=d
        
    def dispP(self):
        print(self.a,self.b,self.c,self.d)

class Child(Parent):
    e=20
    f=50
    
    def __init__(self,g,h,c,d):
        super().__init__(c,d)   #----constructor--chaining----
        self.g=g
        self.h=h
        
    def dispC(self):
        super().dispP()
        print(self.e,self.f,self.g,self.h)
        
ob=Child(100,200,300,400)
ob.dispC()
        
```

```python

def fibonacci(n):
    fib_sequence = [0, 1]
    while len(fib_sequence) < n:
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence[:n]

n = int(input("Enter the number of Fibonacci numbers: "))
print(fibonacci(n))
```

### Shopping Cart Discount(Dictionaries,Loops)


```python
def shopping_cart_discount(cart):
    total = sum(cart.values())
    discounted_total = total
    
    for item, price in cart.items():
        if price > 3000:
            discounted_price = price * 0.95
            discounted_total -= price - discounted_price
    
    if discounted_total > 20000:
        discounted_total *= 0.9
    
    print("Original Total: Rs", total)
    print("Discounted Total: Rs", round(discounted_total, 2))

cart = {
    "laptop": 40000,
    "mouse": 1000,
    "keyboard": 3000
}
shopping_cart_discount(cart)
```
