```python
class GroceryCalculator:
    def __init__(self, items):  # corrected __init__
        self.items = items
        self.quantities = {}
        self.total_cost = 0
        self.itemized_bill = []

    def get_quantities(self):
        print("Enter the quantities for the following items:")
        for item, price in self.items.items():
            quantity_input = input(f"{item.capitalize()} (Price: ${price}/kg): ")
            if quantity_input.isdigit():
                self.quantities[item] = int(quantity_input)
            else:
                print("Invalid input. Please enter a valid number.")
                return False
        return True

    def calculate_total(self):
        for item, quantity in self.quantities.items():
            price = self.items[item]
            cost = quantity * price
            self.total_cost += cost
            self.itemized_bill.append(f"{item.capitalize()}: {quantity}kg x ${price} = ${cost}")

        if self.total_cost > 1000:
            discount = self.total_cost * 0.10
            self.total_cost -= discount
            self.itemized_bill.append(f"Discount Applied: -${discount:.2f}")

    def print_bill(self):
        print("\nItemized Bill:")
        for line in self.itemized_bill:
            print(line)
        print(f"Total Cost: ${self.total_cost:.2f}")

    def run(self):
        if self.get_quantities():
            self.calculate_total()
            self.print_bill()


# Define grocery items and their prices
grocery_items = {"rice": 50, "flour": 30, "sugar": 40}

# Corrected main block
if __name__ == "__main__":
    calculator = GroceryCalculator(grocery_items)
    calculator.run()
```

```python
# 2
def is_strong_password(password):
    upper = lower = digit = special = False
    specials = "!@#$%^&*"

    if len(password) < 8:
        return False

    for ch in password:
        if 'A' <= ch <= 'Z':
            upper = True
        elif 'a' <= ch <= 'z':
            lower = True
        elif '0' <= ch <= '9':
            digit = True
        elif ch in specials:
            special = True

    return upper and lower and digit and special

# Test
print(is_strong_password("SecureP@ss123"))  # True
```

```python
# 3
grades = {
    "Alice": [85, 90, 78],
    "Bob": [72, 81, 84],
    "Charlie": [90, 92, 88]
}

max_avg = 0
topper = ""

for name in grades:
    total = 0
    count = 0
    for score in grades[name]:
        total += score
        count += 1
    avg = total / count
    print(f"{name}: {avg}")
    if avg > max_avg:
        max_avg = avg
        topper = name

print("Topper:", topper)
```

```python
# 4
sentence = input("Enter sentence: ")
sentence = sentence.lower()
cleaned = ""
for ch in sentence:
    if ch.isalpha() or ch == " ":
        cleaned += ch

words = cleaned.split()
freq = {}

for word in words:
    if word not in freq:
        freq[word] = 1
    else:
        freq[word] += 1

print(freq)
```

```python
#5
books = [
    ["Python Basics", "Smith", 5],
    ["Data Science", "Charlie", 3],
    ["Data Structure & Algorithms", "Peter", 2]
]

while True:
    action = input("Borrow or Return? (b/r/exit): ")

    if action == "exit":
        break

    book_name = input("Enter book title: ")
    found = False

    for i in range(len(books)):
        if books[i][0] == book_name:
            found = True
            if action == "b":
                if books[i][2] > 0:
                    books[i][2] -= 1
                    print("Book borrowed.")
                else:
                    print("Out of stock")
            elif action == "r":
                books[i][2] += 1
                print("Book returned.")
    if not found:
        print("Book not found")
```
