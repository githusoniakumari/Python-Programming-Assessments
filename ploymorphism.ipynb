### Hierarchical Inheritance

```python
# Base class
class Grandfather:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def details(self):
        print(f"Grandfather's Name: {self.name}, Age: {self.age}")

# Derived class 1
class Father(Grandfather):
    def __init__(self, name, age, profession):
        super().__init__(name, age)
        self.profession = profession

    def details(self):
        super().details()  # Call Grandfather's details method
        print(f"Father's Profession: {self.profession}")

# Derived class 2
class Son(Father):
    def __init__(self, name, age, profession, hobby):
        super().__init__(name, age, profession)
        self.hobby = hobby

    def details(self):
        super().details()  # Call Father's details method
        print(f"Son's Hobby: {self.hobby}")

# Creating instances of each class
grandfather = Grandfather("Sukhdeb Jha", 105)
father = Father("Pramod Jha", 50, "Manager")
son = Son("Anand", 24, "Manager", "Cricketer")

# Printing details
grandfather.details()
print()
father.details()
print()
son.details()
```

### Polymorphism

### 1. Example of addition

```python
def add(a,b):
    print(a+b)

def add(a,b,c):
    print(a+b+c)
    
#add(10,20)
""" TypeError: add() missing 1 required positional argument: 'c' """
add(10,20,30)
```

### this error is because add(10,20)    ---> method overridden by add(10,20,30)

### Method Overloading can be achieved with the help of Monkey Patching Concept.

```python
def add(a,b):
    print(a+b)
prev=add
def add(a,b,c):
    print(a+b+c)
    
prev(10,20)             
add(10,20,30)
```

```python
class Calculator:
    
    @staticmethod
    def add(a, b):
        return a + b

# Create an object
calc = Calculator()

# Original method
print(calc.add(5, 10))  # Output: 15

# --- Monkey Patching begins ---

# Define a new method (different behavior)
@staticmethod
def add_three_numbers(a, b, c):
    return a + b + c

# Patch the class at runtime
Calculator.add = add_three_numbers

# Now calling the same method name but it works differently
print(calc.add(5, 10, 15))  # Output: 30
```

### PUBLIC PRIVATE PROTECTED ACCESS SPECIFIER

```python

class ClassName:
    def __init__(self):
        self.public_variable = 2

    def public_method(self):
        # method body
        pass

# Object creation and access
obj = ClassName()
print(obj.public_variable)     # Accessing public variable
obj.public_method()            # Calling public method
```

```python
class ClassName:
    def __init__(self):
        self._protected_variable = 2

    def _protected_method(self):
        # method body
        pass

# Object creation and access
obj = ClassName()
print(obj._protected_variable)     # Accessing protected variable
obj._protected_method()            # Calling protected method
```

```python
class ClassName:
    def __init__(self):
        self.__private_variable = 23

    def __private_method(self):
        # method body
        pass

# Object creation and (restricted) access
obj = ClassName()

# ❌ Direct access will raise an AttributeError
# print(obj.__private_variable)         # Error
# obj.__private_method()                # Error

# ✅ Access via name mangling (not recommended for regular use)
print(obj._ClassName__private_variable)
obj._ClassName__private_method()
```

### Abstraction With Real Time Examples 

```python
from abc import ABC, abstractmethod

class ATM(ABC):
    
    @abstractmethod
    def ch_bal(self):
        pass
    
    @abstractmethod
    def withdraw(self):
        pass
    
    @abstractmethod
    def deposit(self):
        pass

class ATM1(ATM):
    def __init__(self, pin):
        self.pin = pin
        self.bal = 0
    
    def ch_bal(self):
        upin = int(input("Enter PIN: "))
        if upin == self.pin:
            print('Available Balance is:', self.bal)
        else:
            print("Incorrect PIN!")

    def deposit(self):
        upin = int(input("Enter PIN: "))
        if upin == self.pin:
            amt = int(input('Enter amount to deposit: '))
            self.bal += amt
            print('Deposit successful.')
            print('Available Balance is:', self.bal)
        else:
            print('Incorrect PIN!')

    def withdraw(self):
        upin = int(input("Enter PIN: "))
        if upin == self.pin:
            amt = int(input('Enter amount to withdraw: '))
            if amt <= self.bal:
                self.bal -= amt
                print('Withdraw successful.')
                print('Available Balance is:', self.bal)
            else:
                print('Insufficient balance!')
        else:
            print('Incorrect PIN!')
            
# Object creation
user1 = ATM1(1234)
user2 = ATM1(5678)
user3 = ATM1(9999)

user1.deposit()
user2.withdraw()
user3.ch_bal()

```
