# bank---atm--user example : MULTI--LEVEL--INHERITANCE

```python
class Bank:
    b_name='Bondhon Bank'
    b_loc='Russia'
    b_ifsc='BND6000'
    
    def __init__(self,name,ph,pin):
        self.name=name
        self.ph=ph
        self.pin=pin
        self.bal=0
    
    
class ATM(Bank):
    atm_loc='Sector V'
    
    
    def ch_bal(self):
        upin=int(input("enter the pin to check the balance: "))
        
        if upin==self.pin:
            print('Available Balance: ',self.bal)
        else:
            print('Invalid Pin:(')
            
    def deposit(self):
        upin=int(input("enter the pin to check the balance: "))
        
        if upin==self.pin:
            amt=int(input("enter the balance to deposit: "))
            
            self.bal+=amt
        else:
            print('Invalid Pin:(')

    def withdraw(self):
        upin = int(input("Enter the pin to withdraw: "))
        if upin == self.pin:
            amt = int(input("Enter the amount to withdraw: "))
            if amt > 0 and amt <= self.bal:
                self.bal -= amt
                print(f'Amount ${amt} is withdrawn from your amount.... Available balance is ${self.bal}')
                print('Thank You visit Again :)')
            elif amt > self.bal:
                print('----Tu Gareeb hai :(----')
            else:
                print('Invalid amount')
        else:
            print('Ghar Jao :(')

            
class User(ATM):
    
    def __init__(self,name,ph,pin):
        super().__init__(name,ph,pin)

ob=User('Dhruv',8867258597,8400)
ob.ch_bal()
```

### REAL TIME EXAMPLE FOR MUTIPLE INHERITANCE

```python
class Brightness:
    def __init__(self, b=0):
        self.b = b

    def turn_off(self):
        self.b = 0
        print('Light is turned off....')

    def turn_on(self):
        self.b = 100
        print('Light is turned on....')

    def set_b(self, new):
        if self.b:
            self.b = new
            print(f'Your brightness is set to {self.b}%')
        else:
            print('Turn on light first...')


class Temperature:
    def __init__(self, temp=35):
        self.temp = temp

    def inc_temp(self):
        self.temp += 1
        print(f'Current temperature is {self.temp}°C')

    def dec_temp(self):
        self.temp -= 1
        print(f'Current temperature is {self.temp}°C')

    def set_temp(self, new):
        self.temp = new
        print(f'Current temperature is {self.temp}°C')


class Thermostat(Brightness, Temperature):
    def __init__(self, b=0, temp=35):
        Brightness.__init__(self, b)
        Temperature.__init__(self, temp)

    def status(self):
        print(f'Current Brightness is {self.b}%')
        print(f'Current Temperature is {self.temp}°C')


# Object creation and method demonstration

# Thermostat object 1
ob1 = Thermostat()
ob1.turn_on()
ob1.set_b(70)
ob1.set_temp(25)
ob1.inc_temp()
ob1.status()

print("\n------------------\n")

# Thermostat object 2
ob2 = Thermostat(b=50, temp=22)
ob2.status()
ob2.turn_off()
ob2.set_b(80)  # Should prompt to turn on light first
ob2.set_temp(20)
```

```python
class Brightness:
    def __init__(self, brightness=0):
        self.brightness = brightness

    def turn_off(self):
        self.brightness = 0
        print('Light is turned off....')

    def turn_on(self):
        self.brightness = 100
        print('Light is turned on....')

    def set_brightness(self, value):
        if self.brightness:
            self.brightness = value
            print(f'Brightness set to {self.brightness}%')
        else:
            print('Turn on the light first...')


class Temperature:
    def __init__(self, temperature=35):
        self.temperature = temperature

    def increase(self):
        self.temperature += 1
        self.show_temperature()

    def decrease(self):
        self.temperature -= 1
        self.show_temperature()

    def set_temperature(self, value):
        self.temperature = value
        self.show_temperature()

    def show_temperature(self):
        print(f'Current temperature is {self.temperature}°C')


class Thermostat(Brightness, Temperature):
    def __init__(self, brightness=0, temperature=35):
        Brightness.__init__(self, brightness)
        Temperature.__init__(self, temperature)

    def status(self):
        print(f'Brightness: {self.brightness}%')
        print(f'Temperature: {self.temperature}°C')


# Usage example
t1 = Thermostat()
t1.turn_on()
t1.set_brightness(75)
t1.set_temperature(24)
t1.increase()
t1.status()
```
