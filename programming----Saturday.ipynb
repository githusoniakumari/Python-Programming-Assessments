### WAP to find product of even digits of a given number


```python
num = int(input("Enter a number: "))

def product_of_even_digits(n):
    product = 1
    has_even = False  # To check if there's at least one even digit
    
    while n > 0:
        digit = n % 10  # get the last digit
        if digit % 2 == 0:  # Check if the digit is even
            product *= digit
            has_even = True
        n //= 10  # Remove the last digit

    if has_even:
        return product
    else:
        return 0  

print("Product of even digits:", product_of_even_digits(num))
```

```python
# 1. Sum of all integers in a given tuple
def sum_of_integers(tup):
    return sum(i for i in tup if isinstance(i, int))



```

```python
# 2. Check if a string is palindrome without slicing
def is_palindrome(s):
    for i in range(len(s) // 2):
        if s[i] != s[-(i + 1)]:
            return False
    return True


```

```python
# 3. Extract strings with length > 4 from a list
def extract_long_strings(lst):
    return [s for s in lst if isinstance(s, str) and len(s) > 4]

```

```python
# 4. Extract uppercase characters at multiples of 3rd index with even ASCII
def extract_uppercase_even_ascii(s):
    return [ch for i, ch in enumerate(s) if i % 3 == 0 and ch.isupper() and ord(ch) % 2 == 0]

```

```python
# 5. Sum of ASCII values of special characters in a string
def sum_special_ascii(s):
    return sum(ord(ch) for ch in s if not ch.isalnum())
```

```python
# 6. Reverse a number without slicing and typecasting
def reverse_number(n):
    rev = 0
    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10
    return rev

```

```python
# 7. Extract palindrome strings from a list
def extract_palindromes(lst):
    return [s for s in lst if isinstance(s, str) and s == s[::-1]]
```

```python
# 8. Remove duplicates from a list without using set
def remove_duplicates(lst):
    unique_list = []
    for i in lst:
        if i not in unique_list:
            unique_list.append(i)
    return unique_list

# Testing the functions
tup = (1, 'hello', 3, 4.5, 5)
print(sum_of_integers(tup))  # Output: 9
```

```python
# Print Factors of a given number

num = int(input("Enter a number: "))

def find_factors(n):
    factors = []
    i = 1
    while i * i <= n:  
        if n % i == 0:
            factors.append(i)
            if i != n // i:  
                factors.append(n // i)
        i += 1
    return sorted(factors)  


print("Factors:", find_factors(num))
```

```python
# wap to print no is prime or not
num = int(input("Enter a number: "))

def is_prime(n):
    if n < 2:
        return False
    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1
    return True

if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
# no is armstrong number or not

num = int(input("Enter a number: "))  
temp = num  
n = len(str(num))
sum_of_powers = 0  

while temp > 0:
    digit = temp % 10  # Get the last digit
    sum_of_powers += digit ** n  # Add digit^n to sum
    temp //= 10  # Remove the last digit

# Check if it's an Armstrong number
if sum_of_powers == num:
    print(f"{num} is an Armstrong number.")
else:
    print(f"{num} is NOT an Armstrong number.")
```

```python
def count_digits(num):
    count = 0
    temp = num
    while temp > 0:
        count += 1
        temp //= 10
    return count

def is_armstrong(num):
    n = count_digits(num)  # Count number of digits
    temp = num
    sum_of_powers = 0
    
    while temp > 0:
        digit = temp % 10  # Extract last digit
        power = 1
        for _ in range(n):  # Compute digit^n without **
            power *= digit
        sum_of_powers += power
        temp //= 10  # Remove last digit
    
    return sum_of_powers == num

# Input from user
num = int(input("Enter a number: "))

# Check and print result
if is_armstrong(num):
    print(f"{num} is an Armstrong number.")
else:
    print(f"{num} is NOT an Armstrong number.")
```

```python
# perfect number or not

def is_perfect(num):
    sum_of_divisors = 0
    
    # Find divisors of num and sum them up
    for i in range(1, num):
        if num % i == 0:
            sum_of_divisors += i
    
    # Check if sum of divisors equals the number
    return sum_of_divisors == num

# Input from user
num = int(input("Enter a number: "))

# Check and print result
if is_perfect(num):
    print(f"{num} is a Perfect number.")
else:
    print(f"{num} is NOT a Perfect number.")


```

```python
# strong numvber or not
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

def is_strong(num):
    temp = num
    sum_of_factorials = 0
    
    while temp > 0:
        digit = temp % 10  # Extract last digit
        sum_of_factorials += factorial(digit)  # Add factorial of digit
        temp //= 10  # Remove last digit
    
    return sum_of_factorials == num

# Input from user
num = int(input("Enter a number: "))

# Check and print result
if is_strong(num):
    print(f"{num} is a Strong number.")
else:
    print(f"{num} is NOT a Strong number.")

```

```python
# spy number or not
def is_spy(num):
    sum_of_digits = 0
    product_of_digits = 1
    temp = num

    while temp > 0:
        digit = temp % 10  # Extract last digit
        sum_of_digits += digit  # Add to sum
        product_of_digits *= digit  # Multiply to product
        temp //= 10  # Remove last digit

    return sum_of_digits == product_of_digits

# Input from user
num = int(input("Enter a number: "))

# Check and print result
if is_spy(num):
    print(f"{num} is a Spy number.")
else:
    print(f"{num} is NOT a Spy number.")

```
